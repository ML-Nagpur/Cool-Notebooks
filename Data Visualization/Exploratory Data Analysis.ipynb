Exploratory data analysis (EDA) is used by data scientists to analyze and investigate data sets and summarize their main characteristics, often employing data visualization methods.
EDA helps determine how best to manipulate data sources to get the answers you need, making it easier for data scientists to discover patterns, spot anomalies, test a hypothesis, or check assumptions.

# Types of exploratory data analysis
### 1. Univariate 
* Data being analyzed consists of just one variable.
* Doesnâ€™t deal with causes or relationships.
### 2. Multivariate 
* More than one variable.
* Shows the relationship between variables of the data through cross-tabulation or statistics.

# Impact of 10th grade on Placements

```python
sns.distplot(df['ssc_p'])
plt.title('Distribution of SSC percentage')
plt.xlabel('SSC %')
```

* The peak of the distribution is around 60-70%, indicating that most students scored within this range.
* The distribution is slightly right-skewed which means fewer students scored very high percentages.
* The data ranges from around 30% to 100%, indicating a wide variation in SSC scores.
* Majority of scores seem to fall between 50% and 80%.
* The right skew suggests that while many students perform around average, high performers are fewer but exist.

## 3. CATPLOT
* Used for categorical plots.
* Shows the relationship between a numerical and one or more categorical variables using one of several visual representations. 

```python
sns.catplot(y='ssc_p', x='status', data=df)
plt.xlabel('Employment Status')
plt.ylabel('SSC %')
```

* Students who are placed have higher SSC percentages, with most scores lying between 60% and 90%.
* Students who are not placed have a wider range of SSC percentages, with most scores between 50% and 80%.
* Placed students mostly have higher SSC scores compared to not placed students.
* The spread for not placed students is slightly wider at the lower end of the SSC percentages, indicating that students with lower SSC percentages are less likely to be placed.
* SSC % is a significant factor in determining employment placement status.
* outliers indicates that while SSC % alone is not an important factor.

## 4. BOXPLOT
* Visualize data distribution for gaining insights and making informed decisions.
* Box plot is a type of chart that depicts a group of numerical data through their quartiles.

```python
sns.boxplot(x='status', y='etest_p', data=df)
plt.title('E-test Scores vs Placement Status')
plt.xlabel('Placement Status')
plt.ylabel('E-test Percentage')
plt.show()
```

* median E-test score for placed students is higher than for not placed students.
* The IQR, which represents the middle 50% of the data, is wider for placed students, indicating a more varied range of E-test scores among those who are placed.
* some students with lower scores also got placed.
* The higher median and IQR for placed students highlight that higher E-test scores generally correlate with better placement outcomes.
* The presence of some placed students with lower scores suggests that while E-test scores are an important factor, they are not the sole determinant of placement.
